```python
symbols = ['EUR/USD', 'USD/JPY', 'GBP/USD', 'USD/CAD', 'USD/CHF', 'AUD/USD', 'NZD/USD', 'EUR/AUD', 'EUR/CAD', 'EUR/CHF', 'EUR/GBP', 'EUR/JPY', 'EUR/NZD', 'GBP/AUD', 'GBP/CAD', 'GBP/CHF', 'GBP/JPY', 'GBP/NZD', 'AUD/CAD', 'AUD/CHF', 'AUD/JPY', 'AUD/NZD', 'NZD/CAD', 'NZD/CHF', 'NZD/JPY', 'CAD/CHF', 'CAD/JPY', 'CHF/JPY']
```

```python
df = dict()
for symbol in symbols:
    df[symbol] = con.get_candles(symbol, period='m5', number=5760) #2 Days
    df[symbol]['Open'] = (df[symbol]['bidopen'] + df[symbol]['askopen']) / 2
    df[symbol]['High'] = (df[symbol]['bidhigh'] + df[symbol]['askhigh']) / 2
    df[symbol]['Low'] = (df[symbol]['bidlow'] + df[symbol]['asklow']) / 2
    df[symbol]['Close'] = (df[symbol]['bidclose'] + df[symbol]['askclose']) / 2
    df[symbol].drop(['bidopen', 'askopen', 'bidhigh', 'askhigh', 'bidlow', 'asklow', 'bidclose', 'askclose', 'tickqty'],axis=1, inplace=True)
```

```python
log_returns = pd.DataFrame()
```

```python
for symbol in symbols:
    log_returns[symbol] = np.log(df[symbol]['Close'] / df[symbol]['Close'].shift(1))
```

```python
cov_matrix = log_returns.cov()
cov_matrix
```
